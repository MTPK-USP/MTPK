```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P0_FKP = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_P0_FKP.dat')
P0_MT = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_P0_MTOE.dat')

P2_FKP = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_P2_FKP.dat')
P2_MT = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_P2_MTOE.dat')

P4_FKP = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_P4_FKP.dat')
P4_MT = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_P4_MTOE.dat')

C0_FKP = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_C0_FKP.dat')
C2_FKP = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_C2_FKP.dat')
C4_FKP = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_C4_FKP.dat')

k = np.loadtxt('spectra/newExSHalos/k=0.1/newExSHalos_vec_k.dat')
```

```python
P0_FKP.shape, P0_MT.shape, P2_FKP.shape, P2_MT.shape, C0_FKP.shape, C2_FKP.shape, C4_FKP.shape, k.shape
```

```python
P0_FKP = P0_FKP.reshape(3, 4, 40)
P0_MT = P0_MT.reshape(3, 4, 40)

P2_FKP = P2_FKP.reshape(3, 4, 40)
P2_MT = P2_MT.reshape(3, 4, 40)

P4_FKP = P4_FKP.reshape(3, 4, 40)
P4_MT = P4_MT.reshape(3, 4, 40)

C0_FKP = C0_FKP.reshape(3, 4, 40)
C2_FKP = C2_FKP.reshape(3, 4, 40)
C4_FKP = C4_FKP.reshape(3, 4, 40)
```

```python
P0_FKP.shape
```

```python
np.mean(P0_FKP, axis = 1).shape
```

```python
plt.figure(dpi = 100)
for i in range(3):
    plt.loglog(k, np.mean(P0_FKP, axis = 1)[i], label = f'FKP Monopole - Bin {i}')

    plt.loglog(k, np.mean(P0_MT, axis = 1)[i], '--', label = f'MT Monopole - Bin {i}')
    plt.legend()
```

```python
plt.figure(dpi = 100)
for i in range(3):
    plt.plot(k, np.mean(P2_FKP, axis = 1)[i], label = f'FKP Dipole - Bin {i}')

    plt.plot(k, np.mean(P2_MT, axis = 1)[i], '--', label = f'MT Dipole - Bin {i}')
    plt.legend()
```

```python
plt.figure(dpi = 100)
for i in range(3):
    plt.plot(k, np.mean(P4_FKP, axis = 1)[i], label = f'FKP Quadrupole - Bin {i}')

    plt.plot(k, np.mean(P4_MT, axis = 1)[i], '--', label = f'MT Quadrupole - Bin {i}')
    plt.legend()
```

```python
plt.figure(dpi = 100)
for i in range(3):
    plt.loglog(k, np.mean(C0_FKP, axis = 1)[i], label = f'Cross Monopole - Bin {i}')
    plt.legend()
```

```python
plt.figure(dpi = 100)
for i in range(3):
    plt.plot(k, np.mean(C2_FKP, axis = 1)[i], label = f'Cross Dipole - Bin {i}')
    plt.xscale('log')
    plt.legend()
```

```python
plt.figure(dpi = 100)
for i in range(3):
    plt.plot(k, np.mean(C4_FKP, axis = 1)[i], label = f'Cross Quadrupole - Bin {i}')
    plt.xscale('log')
    plt.legend()
```
